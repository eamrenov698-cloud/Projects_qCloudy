# 2. Заполнение пропущенных значений

```python
df.isna().sum()
```

```python
from openai import OpenAI
import os
import time
import numpy as np

def fill_empty_cells_with_gpt(df, api_key, context=""):
    df_filled = df.copy()
    client = OpenAI(api_key=api_key)
    
    for column in df.columns:
        for index in df.index:
            if pd.isna(df.loc[index, column]):
                row_context = df.loc[index].to_dict()
                
                prompt = f"""На основе следующих данных строки: {row_context}
                и дополнительного контекста: {context}
                предложите подходящее значение для пустой ячейки в столбце '{column}'.
                Дайте только значение, без объяснений."""
                
                try:
                    response = client.chat.completions.create(
                        model="gpt-3.5-turbo",
                        messages=[
                            {"role": "system", "content": "Вы - помощник по заполнению пропущенных данных. Отвечайте только значением, без объяснений."},
                            {"role": "user", "content": prompt}
                        ],
                        temperature=0.7
                    )
                    
                    suggested_value = response.choices[0].message.content.strip()
                    
                    if df[column].dtype == 'int64':
                        suggested_value = int(suggested_value)
                    elif df[column].dtype == 'float64':
                        suggested_value = float(suggested_value)
                    
                    df_filled.loc[index, column] = suggested_value
                    time.sleep(0.5)
                    
                except Exception as e:
                    print(f"Ошибка при заполнении ячейки [{index}, {column}]: {str(e)}")
                    continue
    
    return df_filled


print("Исходный DataFrame:")
print(df)
print("\nПоиск пустых значений:")
print(df.isna().sum())

api_key = "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"
context = "Это данные о ёлках, там есть такие колонки как: вес, подставка, ветви. И в некоторых колонках есть пропущенные значения, нужно их заполнить"

df_filled = fill_empty_cells_with_gpt(df, api_key, context)

print("\nЗаполненный DataFrame:")
print(df_filled)
```

```python
df_dr['Вес'] = df_dr['Вес'].str.rstrip('.')
df_dr['Вес'] = df_dr['Вес'].str.rstrip().str.lstrip()
df_dr['Вес'] = df_dr['Вес'].str.replace(' ', '')
df_dr['Вес'] = df_dr['Вес'].str.replace('kg', '')
df_dr['Вес'] = df_dr['Вес'].str.replace('К', '')
```

```python
def convert_to_kg(value):
    if isinstance(value, (int, float)):
        return value
    value = str(value)  
    
    if 'кг' in value:
        return float(value.replace('кг', '').strip())
        
    elif 'г' in value:
        return float(value.replace('г', '').strip()) / 1000
        
    else:
        return value

df_dr['Вес'] = df_dr['Вес'].apply(convert_to_kg)
```

```python
df_dr['Вес'] = df_dr['Вес'].astype(float)
df_dr['Вес'].dtype
```
